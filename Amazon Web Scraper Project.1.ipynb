```python
# to automate and collect data in the background 

def check_price():
    URL = 'https://www.amazon.co.uk/Samsung-Smartphone-Extended-Manufacturer-Warranty/dp/B0BQRNZNY6/ref=sr_1_2?crid=1TB892FKK4WL3&keywords=samsung%2Bgalaxy%2Bs23%2Bultra&qid=1703256852&sprefix=samung%2Bgalaxy%2Bs23%2Bultra%2Caps%2C350&sr=8-2&th=1'

    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36"}

    page = requests.get(URL, headers=headers)

    soup1 = BeautifulSoup(page.content, 'html.parser')

    soup2 = BeautifulSoup(soup1.prettify(), 'html.parser')

    title = soup2.find(id='productTitle').get_text()
    price = soup2.find(id='corePrice_feature_div').get_text()

    review = soup2.find(id='acrCustomerReviewText').get_text()
    price = price.strip()[1:]
    title = title.strip()
    review = review.strip()
    
    import datetime
    today = datetime.date.today()
    
    
    import csv
    header = ['Title', 'Price', 'Review', 'Date']
    data = [title, price, review, today]
    
    
    with open('AmazonWebScraperDataset.csv', 'a+', newline='', encoding='UTF8') as f:
        writer = csv.writer(f)
        writer.writerow(data)
        
        
        
    if(price < 1299):
        send_mail()
    
    
    
```

```python
# o collect data after 5 seconds

while(True):
    check_price()
    time.sleep(5)
```

```python
# to collect data after 24hrs

while(True):
    check_price()
    time.sleep(86400)
```
